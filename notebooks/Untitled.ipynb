```python
block_size = 100
xs = np.arange(1, 21)
i = (np.log(block_size) / block_size * np.arange(1, 21)).round(2)

idx = np.arange(len(xs))
mask = np.zeros((len(xs), len(xs)), dtype=bool)

for a, b in product(xs, xs):
    if a >= b:
        mask[a-1, b-1] = (np.abs(np.sqrt(a) - np.sqrt(b)) >= np.sqrt(2))
pd.DataFrame(mask, columns=i, index=i).T.sort_index(ascending=True).sort_index(axis=1, ascending=True)
```

```python
n_samples=100
n_features=10
n_mindsets=10
range_answers = [1, 10
                ]
seed = 42
```

```python
min_answer = range_answers[0]
max_answer = range_answers[1]
if seed is not None:
    np.random.seed(seed)

xs, ys = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_mindsets)
```

```python
for i in np.unique(ys):
    xs_mindset = xs[ys == i]
    xs_mindset = np.interp(xs_mindset, (xs_mindset.min(), xs_mindset.max()), (min_answer, max_answer))
    xs[ys == i] = np.floor(xs_mindset).astype(int)

xs = xs.astype(int)
```

```python
xs_reduced = TSNE(n_components=2).fit_transform(xs)
plt.scatter(xs_reduced[:, 0], xs_reduced[:, 1], c=ys)
```

```python
colums_name = [f'q{i}' for i in range(1, n_features+1)]
df = pd.DataFrame(xs, columns=colums_name)
cut_values = np.arange(min_answer+1, max_answer+1)
nb_points = df.shape[0]
questions = []

df_binarized = pd.DataFrame()
for column in df.columns:
    for cut_value in cut_values:
        new_col = np.zeros(nb_points, dtype=bool)
        new_col[df[column] < cut_value] = 0
        new_col[df[column] >= cut_value] = 1

        short_name = f'{column}_{cut_value}-{max_answer}'
        questions.append(f'{column} larger than {cut_value}')

        df_binarized[short_name] = new_col
```

```python
import re
template = re.compile(r"(\w+) .+ (\d)")
```

```python
def change_lower(interval, new_value):
    if new_value > interval[0]:
        return (new_value, interval[1])

    return interval

def change_upper(interval, new_value):
    if new_value < interval[1]:
        return (interval[0], new_value)

    return interval
```

```python
df = pd.read_csv('../output/0/q/best.csv')
nb_tangles = len(df)
inteval_values = (1, 5)
results = {}

for column in df.columns:
    
    name, value = template.findall(column)[0]
    value = int(value)
    
    old = results.get(name, None)
    if old is None:
        new = [inteval_values for _ in range(nb_tangles)]
    else:
        new = old
        
    for i, take in enumerate(df[column]):
        if take:
            new[i] = change_lower(new[i], value)
        else:
            new[i] = change_upper(new[i], value-1)
    results[name] = new

a = pd.DataFrame.from_dict(results)
f = lambda i: i if i[0] != i[1] else i[0]
a = a.applymap(f)
```
